```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
```

```python
df1 = sns.load_dataset('titanic')
df1.head()
```

```python
df1.isnull().mean()*100
```

```python
df = df1.drop('deck', axis=1)
df.head()
```

```python
df.isnull().mean()*100
```

```python
df.info()
```

```python
df['adult_male'].replace(True, 1, inplace = True)
df['adult_male'].replace(False, 0, inplace = True)
df['alive'].replace('yes', 1, inplace = True)
df['alive'].replace('no', 0, inplace = True)
df['alone'].replace(True, 1, inplace = True)
df['alone'].replace(False, 0, inplace = True)
```

```python
x = df.drop('survived', axis=1)
y = df.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)

X_train.shape, X_test.shape
y
```

```python
x.info()
```

```python
# filling
trf1 = ColumnTransformer([
        ('imputer', SimpleImputer(strategy='mean'),[2]),          # us the index becasue it return the numpy array
        ('imputer1', SimpleImputer(strategy='most_frequent'), [6,10])
] , remainder='passthrough')
trf1
```

```python
# encoding
trf2 = ColumnTransformer([
    ('encoder', OneHotEncoder(sparse_output=False, drop='first'), [1,2,4,9]),
    ('encoder_1', OrdinalEncoder(categories=[['First','Second','Third']]),[8])
], remainder='passthrough')
trf2
# trf2.fit(X_train)              # to find the number of column
# len(trf2.get_feature_names_out())
```

```python
# # another method to find no. of columns after the encoding
# X_train_trf2 = trf2.fit_transform(X_train)

# print(X_train_trf2.shape)
```

```python
# scaling
trf3 = ColumnTransformer([
    ('scale', MinMaxScaler(), slice(0,16))
], remainder='passthrough')
```

```python
# train the model
trf5 = DecisionTreeClassifier()
```

```python
# creating pipeline

pipe = Pipeline(
    [
        ('trf1', trf1),
        ('trf2', trf2),
        ('trf3', trf3),

        ('trf5', trf5)
    ]
)
```

```python
pipe.fit(X_train, y_train)
```
