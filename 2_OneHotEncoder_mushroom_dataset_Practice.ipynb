```python
x = df.drop(columns='class')
y = df.iloc[:,0]

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape, y_train.shape
```

```python
ohe = OneHotEncoder(sparse_output = False, drop='first')

X_train_new = ohe.fit_transform(X_train)
X_test_new = ohe.transform(X_test)
```

```python
lb = LabelEncoder( )

y_train_new = lb.fit_transform(y_train)
y_test_new = lb.transform(y_test)
```

```python
lg = LogisticRegression()

lg.fit(X_train_new, y_train_new)

y_pred = lg.predict(X_test_new)

accuracy_score(y_test_new, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
lg = LogisticRegression()
scores = cross_val_score(lg, X_train_new, y_train_new, cv=5)

print(scores)
print(scores.mean())
```
