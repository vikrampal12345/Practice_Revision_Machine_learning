# 📌 Pandas `pd.cut()`

## 📖 Definition
`pd.cut()` is a pandas function used to convert continuous numerical data into categorical intervals (bins). This process is known as **Binning** or **Discretization**.

---

## 🎯 Why do we use `pd.cut()`?

Suppose we have earthquake magnitudes:

| Magnitude |
|-----------|
| 4.3 |
| 5.4 |
| 6.5 |
| 7.8 |

Instead of using continuous values, we can convert them into categories:

| Magnitude | Category |
|-----------|----------|
| 4.3 | Moderate |
| 5.4 | Strong |
| 6.5 | Very Strong |
| 7.8 | High |

This makes the data easier to understand and is useful in **EDA** and **Feature Engineering**.

---

# 📝 Syntax

```python
pd.cut(
    x,
    bins,
    labels=None,
    right=True
)
```

---

# ⚙️ Parameters

### **x**
The numerical column that you want to divide into categories.

```python
df['Magnitude']
```

---

### **bins**
Defines the interval boundaries.

```python
bins = [0, 5, 6, 7, float('inf')]
```

Intervals become:

```
0 -------- 5 -------- 6 -------- 7 -------- ∞
```

---

### **labels**
Names assigned to each interval.

```python
labels = [
    'Moderate',
    'Strong',
    'Very Strong',
    'High'
]
```

---

### **right**

Determines which side of the interval is included.

### **right=True (Default)**

```
(0,5]
(5,6]
(6,7]
```

Meaning:

```
0 < x ≤ 5
```

Example:

```
5 → Moderate
6 → Strong
7 → Very Strong
```

---

### **right=False**

```
[0,5)
[5,6)
[6,7)
```

Meaning:

```
0 ≤ x < 5
```

Example:

```
5 → Strong
6 → Very Strong
7 → High
```

---

# 💻 Example

```python
import pandas as pd

bins = [0,5,6,7,float('inf')]

labels = [
    'Moderate',
    'Strong',
    'Very Strong',
    'High'
]

df['Earthquake_Category'] = pd.cut(
    df['Magnitude'],
    bins=bins,
    labels=labels,
    right=False
)
```

---

# 📊 Output

| Magnitude | Earthquake_Category |
|-----------|---------------------|
| 4.3 | Moderate |
| 5.5 | Strong |
| 6.4 | Very Strong |
| 7.9 | High |

---

# ⭐ Important Points

- `pd.cut()` is used for **Binning**.
- `bins` define the interval boundaries.
- `labels` assign names to each interval.
- `right=True` → Right endpoint included.
- `right=False` → Left endpoint included.
- `float('inf')` means **positive infinity**.

---

# ⚠️ Common Mistakes

❌ Number of labels must always be **one less** than the number of bin boundaries.

```python
bins = [0,5,6,7]
```

Here,

```
Intervals = 3
```

So labels must also be **3**.

---

# 💡 Memory Trick

```
Continuous Values
        ↓
    pd.cut()
        ↓
Categories (Bins)
```

# 🎤 Interview Question

### Q. What is the difference between `pd.cut()` and `pd.qcut()`?

**Answer:**

- `pd.cut()` creates intervals using **fixed boundaries**.
- `pd.qcut()` creates intervals using **equal number of observations**.
